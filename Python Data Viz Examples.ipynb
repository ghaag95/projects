```python
#ML with Python Assignment 3 Question 1
#Glenn Haag 06/13/2021

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#import the csv file
data = pd.read_csv('barGraph.csv')
#set the width of each bar to .1
barWidth = .1
#set the plot to be a 10x10
plt.subplots(figsize = (10,10))

#create each bar's x coordinates
anticipation = np.arange(len(data.loc[:, "Business"]))
enjoyment = [x + barWidth for x in anticipation]
sad = [x + barWidth for x in enjoyment]
disgust = [x + barWidth for x in sad]
anger = [x + barWidth for x in disgust]
surprise = [x + barWidth for x in anger]
fear = [x + barWidth for x in surprise]
trust = [x + barWidth for x in fear]

#plot the bars with the data
plt.bar(anticipation, data.loc[:, 'anticipation'], color = 'b', align = 'edge', width = barWidth, label = 'Anticipation')
plt.bar(enjoyment, data.loc[:, 'enjoyment'], color = 'orange', align = 'edge', width = barWidth, label = 'Enjoyment')
plt.bar(sad, data.loc[:, 'sad'], color = 'g', align = 'edge', width = barWidth, label = 'Sad')
plt.bar(disgust, data.loc[:, 'disgust'], color = 'r', align = 'edge', width = barWidth, label = 'Disgust')
plt.bar(anger, data.loc[:, 'anger'], color = 'm', align = 'edge', width = barWidth, label = 'Anger')
plt.bar(surprise, data.loc[:, 'surprise'], color = 'y', align = 'edge', width = barWidth, label = 'Surprise')
plt.bar(fear, data.loc[:, 'fear'], color = 'k', align = 'edge', width = barWidth, label = 'Fear')
plt.bar(trust, data.loc[:, 'trust'], color = 'pink', align = 'edge', width = barWidth, label = 'Trust')

#label the axes
plt.xlabel('Emotion')
plt.ylabel('Number of respondents')

#space out the x axis ticks and rotate for readability
plt.xticks([r + barWidth for r in range(len(data.loc[:, "Business"]))],
        data.loc[:, "Business"], rotation = 45)
#display legend and plot
plt.legend()
plt.show()




```

```python
#ML with Python Assignment 3 Question 2
#Glenn Haag 06/13/2021

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#import the csv file
data = pd.read_csv('barGraph.csv')
#create 6 different sub plots for the 6 pie charts
fig, axs = plt.subplots(nrows = 3, ncols = 2, figsize = (15,15))
#create the labels for the legend, but take out 'Business'
myLabels = data.columns.to_list()
myLabels.remove('Business')
#counter for the row of the data we want the data from
n = 0
#for each subplot
for x in range(3):
    for y in range(2):
        #plot the pie charts with the data in the data frame and display percents
        axs[x][y].pie(data.iloc[n, data.columns != 'Business'], autopct='%1.1f%%')
        #title each pie chart
        axs[x][y].set_title(data['Business'][n])
        #set the legend to display on each subplot
        axs[x][y].legend(loc = 1, bbox_to_anchor = (1.2, 1.2), labels = myLabels, prop = {'size': 8})
        n += 1
#display the pie charts        
plt.tight_layout()
plt.show()

```

```python
#ML with Python Assignment 3 Question 3
#Glenn Haag 06/13/2021

import os
import matplotlib.pyplot as plt
from wordcloud import WordCloud

d = os.getcwd()

filepath=d+'\obamaSpeech.txt'


def wordCloud(path):
    #open the speech text file
    text = open(path).read() 
    #generate the wordCloud
    wordcloud = WordCloud().generate(text)
    #set the minimum word length to 4 and make it more readable
    wordcloud = WordCloud(min_word_length = 4,background_color="white", max_words=2000,max_font_size=40, relative_scaling=.4).generate(text)
    #make the figure size bigger to improve readability
    plt.figure(figsize = (12,12), dpi = 80)
    #display the wordCloud
    plt.imshow(wordcloud)
    #disable the axis
    plt.axis("off")

wordCloud(filepath)
```

```python
#ML with Python Assignment 3 Question 4
#Glenn Haag 06/13/2021

import os
import matplotlib.pyplot as plt


d = os.getcwd()

filepath=d+'\ReviewID.txt'

text = open(filepath).read()
#make a dictionary of all the languages
langDict = {}
#for each line of the text file
for line in text.split('\n'):
    #get the language (characters after the colon)
    #I deleted the blank line at the bottom of the text file to avoid having an (if line not empty) clause
    lang = line.split(':')[1]
    #if the language isnt in the dictionary, add it, else incrememnt the count of the language
    if lang not in langDict:
        langDict[lang] = 1
    else:
        langDict[lang] += 1
        
#create bar graph using the data from the dictionary I created
graph = plt.bar(langDict.keys(), langDict.values())
plt.show()

    
```
